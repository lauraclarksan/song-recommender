```python
def get_playlist_tracks(playlist_id):

    results = sp.user_playlist_tracks("spotify", playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("5XlfGz8AhCxlWq0zSGF6Mn")
tracks
```

```python
track_ids = [track['track']['uri'] for track in tracks]
track_ids
```

```python
def get_artists_and_ids_from_playlist(playlist_id):

    tracks_from_playlist = get_playlist_tracks(playlist_id)

    artists = []
    artists_ids = []

    for track in tracks_from_playlist:
        artists_info = track['track']['artists']

        for artist_info in artists_info:
            artists.append(artist_info['name'])
            artists_ids.append(artist_info['id'])

    return list(set(artists)), list(set(artists_ids))

artists, artists_ids = get_artists_and_ids_from_playlist("5XlfGz8AhCxlWq0zSGF6Mn")
artists_ids
```

```python
start = 0
df = pd.DataFrame()

for stop in range(0, len(track_ids)+100, 100):
    if start != stop:
        print(start, stop)
        new_df = pd.json_normalize(sp.audio_features(track_ids[start:stop]))
        df = pd.concat([new_df, df])
        start = stop

df.reset_index()
```

```python
track_features = df.copy()
```

```python
track_features
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
X = track_features[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness','acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']]

X
```

```python
from sklearn.preprocessing import StandardScaler

X_prep = StandardScaler().fit_transform(X)

pd.DataFrame(X_prep).head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)  #selected cluster = 3 after using the Silhouette Scores 
kmeans.fit(X_prep)
```

```python
clusters = kmeans.predict(X_prep)
clusters
```

```python
pd.Series(clusters).value_counts().sort_index()
```

```python
X_df = pd.DataFrame(X)
X_df["cluster"] = clusters
```

```python
cluster_mapping = {i: i for i in range(3)}
```

```python
X_df["cluster"] = X_df["cluster"].apply(lambda x: cluster_mapping[x])
X_df
```
